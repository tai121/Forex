```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('GBPUSD_Candlestick_1_D_ASK_01.01.2010-20.05.2023.csv')
df.info()
```

```python
df['Gmt time'] = df['Gmt time'].astype("string").str.split(' ').str[0]
df.head(30)
```

```python
df['Gmt time'] = pd.to_datetime(df['Gmt time'],format='%d.%m.%Y')
df
```

```python
df.columns = ['Date','Open','High','Low','Close','Volume']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x='Date', y='Close', data=df)
plt.xlabel('date')
plt.ylabel('close')
plt.title('Line Chart: Date vs Close')
plt.show()
```

```python
sns.lineplot(x='Date', y='Volume', data=df)
plt.xlabel('date')
plt.ylabel('volume')
plt.title('Line Chart: Date vs volume')
plt.show()
```

```python
df['day_in_week'] = df['Date'].dt.dayofweek
```

```python
df['Change'] = abs(df.High - df.Low)
```

```python
for i in range(7):
    print(df[df['day_in_week'] == i]['Volume'].mean(),df[df['day_in_week'] == i]['Change'].max())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df['close_nor'] = scaler.fit_transform(df[['Close']])
```

```python
scaler = MinMaxScaler()
df['volume_nor'] = scaler.fit_transform(df[['Volume']])
```

```python
scaler = MinMaxScaler()
df['Change_nor'] = scaler.fit_transform(df[['Change']])
df['VolumeChange'] = df['Volume'] - df['Volume'].shift(-1)
scaler = MinMaxScaler()
df['VolumeChange_nor'] = scaler.fit_transform(df[['VolumeChange']])
```

```python
UK = pd.read_csv('UK_rate.csv')
UK.columns = ['Date','Rate']
UK['Date'] = pd.to_datetime(UK['Date'])
UK
```

```python
US = pd.read_csv('DFF.csv')
US.columns = ['Date','Rate']
US['Date'] = pd.to_datetime(US['Date'],format='%d/%m/%Y')
US.head()
```

```python
US = US[US['Date'] <= '2023-05-19']
```

```python
df['UK_interest_rate'] = 0
df['US_interest_rate'] = 0
```

```python
i = j = 0
while i<len(df) and j< len(UK):
    if df.iloc[0]['Date']== UK.iloc[0]['Date']:
        df.loc[i,'UK_interest_rate'] = UK.loc[i,'Rate']
        i+=1
        j+=1
    else:
        j+=1
```

```python
i = j = 0
while i<len(df) and j< len(US):
    if df.iloc[0]['Date']== US.iloc[0]['Date']:
        df.loc[i,'US_interest_rate'] = US.loc[i,'Rate']
        i+=1
        j+=1
    else:
        j+=1
```

```python
scaler = MinMaxScaler()
df['change_nor'] = scaler.fit_transform(df[['change']])
```

```python
df
```

```python
scaler = MinMaxScaler()
df['change_in_day'] = df['Change'].diff()
df['changeinday'] = scaler.fit_transform(df[['change_in_day']])
```

```python
import plotly.graph_objects as go
fig = go.Figure()
x = 0
y = 60
# ThÃªm line chart cho cá»™t 'Close'
fig.add_trace(go.Scatter(x=df['Date'][x:y], y=df['Change_nor'][x:y], name='Close'))

# ThÃªm line chart cho cá»™t 'Volume'
fig.add_trace(go.Scatter(x=df['Date'][x:y], y=df['volume_nor'][x:y], name='Volume'))

# Äáº·t tÃªn trá»¥c x vÃ  y
fig.update_layout(xaxis_title='Date', yaxis_title='Value')

# Hiá»ƒn thá»‹ biá»ƒu Ä‘á»“
fig.show()
```
